# Here we have some columns with null values

## we fill these values with their mean,meadian, mode (most common values)... 

```python
df['agent']=df['agent'].fillna(df['agent'].mean())
```

```python
most_fre_country=df['country'].mode()[0]
df['country']=df['country'].fillna(most_fre_country)
most_fre_country=df['company'].mode()[0]
```

```python
df['company']=df['company'].fillna(most_fre_country)
df['children']=df['children'].fillna(df['children'].mean())
```

# Cancellation Rate on Resort 

## we normalize the values in perecent(%) format to make it easy by using big count values 

```python
resort_hotel=df[df['hotel']=='Resort Hotel']
# resort_hotel['is_canceled'].value_counts()
cancellation_rate=resort_hotel['is_canceled'].value_counts(normalize=True)
cancellation_rate
```

```python
ax=sns.barplot(x=cancellation_rate.index,y=cancellation_rate.values)
plt.title("Cancellation rate for Resort hotel",size=15,color='r')
for bars in ax.containers:
    ax.bar_label(bars)
plt.ylabel('Proprtion')
plt.xlabel("Is Canceled")
plt.xticks([0,1],['Not canceled','Canceled'])
```

# line charts for both city and resort hotels, showing the trend of the average daily rate (ADR) per month

### Filter the DataFrame for canceled reservations in city hotel, group by month, sum adr, and reset the index


### Filter the DataFrame for canceled reservations in resort hotel, group by month, sum adr, and reset the index


```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 8))
plt.title("ADR per month",fontweight='bold',fontsize=15)

# Filter the DataFrame for canceled reservations in city hotel, group by month, sum adr, and reset the index
city_canceled_by_month = df[(df['is_canceled'] == 1) & (df['hotel'] == 'City Hotel')].groupby('month')['adr'].sum().reset_index()
plt.plot(city_canceled_by_month['month'], city_canceled_by_month['adr'], label='City Hotel', marker='o', linestyle='-')

# Filter the DataFrame for canceled reservations in resort hotel, group by month, sum adr, and reset the index
resort_canceled_by_month = df[(df['is_canceled'] == 1) & (df['hotel'] == 'Resort Hotel')].groupby('month')['adr'].sum().reset_index()

plt.plot(resort_canceled_by_month['month'], resort_canceled_by_month['adr'], label='Resort Hotel', marker='*', linestyle='-')

plt.xlabel('Month')
plt.ylabel('ADR')

plt.legend()

plt.grid(True)  # Add gridlines for better readability

plt.show()
```
